```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import sys
```

```python
model_id = "/data/usr/jy/Langchain-Chatchat/webui_pages/fine_tune/final_model/user_模型2401"
tokenizer = AutoTokenizer.from_pretrained(model_id,trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_id)
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=10)
input_text="模具是什么？"
chunk_size=100
input_chunks = [input_text[i:i+chunk_size] for i in range(0, len(input_text), chunk_size)]
```

```python
def generate_streaming_output(model_path, input_text, chunk_size=50):
    # 加载分词器和模型
    tokenizer = AutoTokenizer.from_pretrained(model_path,trust_remote_code=True)
    model = AutoModelForCausalLM.from_pretrained(model_path)

    # 初始化文本生成的 pipeline
    text_generator = pipeline("text-generation", model=model, tokenizer=tokenizer, device=0)

    # 将输入文本分割为较小的块
    input_chunks = [input_text[i:i+chunk_size] for i in range(0, len(input_text), chunk_size)]
    
    for chunk in input_chunks:
        # 对每个块生成文本
        output = text_generator(chunk, max_length=10, return_full_text=False)
        generated_chunk = output[0]['generated_text'].replace(chunk, '').strip()
        yield generated_chunk  # 使用 yield 逐块产生输出
```

```python
for generated_chunk in generate_streaming_output(model_id, "模具是什么，输出1000字"):
    print("Generated chunk:", generated_chunk)
    sys.stdout.flush()
```
